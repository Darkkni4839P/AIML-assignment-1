```python
# question 2
def calculate_average(values):
  total = 0
  length = 0

  for i in range(len(values)):
    total += values[i]
    length += 1
  if length == 0:
    return None
  return total / length
numbers = [5, 15, 25, 35, 45, 55]
avg = calculate_average(numbers)
if avg is not None:
  print("average:", avg)
else:
  print("list is empty, average cannot be calculated")



```

```python
# question 5
def mean_median_mode(numbers):
    if len(numbers) == 0:
        print("Error: Empty list")
        return
    total = 0
    for num in numbers:
        total += num
    mean = total / len(numbers)
    numbers.sort()
    n = len(numbers)
    if n % 2 == 0:
        median = (numbers[n//2 - 1] + numbers[n//2]) / 2
    else:
        median = numbers[n//2]
    frequency = {}
    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1
    max_freq = max(frequency.values())
    mode = []
    for key in frequency:
        if frequency[key] == max_freq and max_freq > 1:
            mode.append(key)
    if not mode:
        mode = "No mode"
    print("Mean:", mean)
    print("Median:", median)
    print("Mode:", mode)
mean_median_mode([1.6, 2, 2, 3, 3, 4.9])
```

```python
# question 6
values = [1.25, 3.5, 5.75, 7.25, 9.75]
if not values:
    print("List is empty")
else:
    n = 0
    total = 0
    for x in values:
        total += x
        n += 1
    mean = total / n
    total_squared_diff = 0
    for x in values:
        diff = x - mean
        total_squared_diff += diff * diff
    std_dev = (total_squared_diff / n) ** 0.5
    print("Standard Deviation:", std_dev)
```

```python
# question 7
A = [
    [1, 2, 3],
    [4, 5, 6]
]
B = [
    [7, 8, 9],
    [1, 2, 3]
]
if len(A) != len(B) or len(A[0]) != len(B[0]):
    print("Matrix dimensions do not match")
else:
    C = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        C.append(row)
    print("Resultant Matrix:")
    for row in C:
        print(row)
```

```python
# question 8
actual = [3, 5.2, 2, 7.7]
predicted = [2.1, 5, 4.7, 6]
if len(actual) != len(predicted):
    print("Error: Lists must be of the same length")
else:
    n = len(actual)
    abs_error_sum = 0
    sq_error_sum = 0
    for i in range(n):
        error = actual[i] - predicted[i]
        abs_error_sum += abs(error)
        sq_error_sum += error * error
    mae = abs_error_sum / n
    mse = sq_error_sum / n
    print("MAE:", mae)
    print("MSE:", mse)

```

```python
# question 9
matrix = [
    [1, 2, 3],
    [4, 5.5, 6],
    [7, 8, 9]
]
rows = len(matrix)
cols = len(matrix[0])
if rows != cols:
    print("Error: Matrix is not square")
else:
    if rows == 2:
        det = matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]
        print("Determinant:", det)
    elif rows == 3:
        det = (
            matrix[0][0]*(matrix[1][1]*matrix[2][2] - matrix[1][2]*matrix[2][1])
            - matrix[0][1]*(matrix[1][0]*matrix[2][2] - matrix[1][2]*matrix[2][0])
            + matrix[0][2]*(matrix[1][0]*matrix[2][1] - matrix[1][1]*matrix[2][0])
        )
        print("Determinant:", det)
    else:
        print("Only 2x2 or 3x3 matrices are supported")
```
